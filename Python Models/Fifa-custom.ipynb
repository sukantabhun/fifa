# Functions

```python
def Player_Score(Name,Forwards):
    data = Forwards[Forwards['Name']==Name]
    data = data.loc[:,'Crossing':'SlidingTackle'].select_dtypes(exclude = 'object')
    score = data.sum().sum()
    return round(score/29,2)
```

```python
def GK_Score(Name, goal_keepers):
    data = goal_keepers[goal_keepers['Name'] == Name]
    data = data.loc[:, 'GKDiving':'GKReflexes'].select_dtypes(exclude='object')
    score = data.sum().sum()
    return round(score/5, 2)
```

# Data Manupulations

```python
player_data['Best Position']=player_data['Best Position'].replace(['CB','RB','LB','RWB','LWB'],'Back')
```

```python

player_data['Best Position']=player_data['Best Position'].replace('GK','Goal Keeper')
```

```python
player_data['Best Position']=player_data['Best Position'].replace(['CAM','CM','CDM','LM','RM'],'Mid')
```

```python
player_data['Best Position']=player_data['Best Position'].replace(['ST','LW','RW','CF'],'Forwards')
```

# TOP 20 Goal keepers

```python
goal_keepers = player_data[player_data['Best Position'] == 'Goal Keeper']
GKList = pd.DataFrame(columns=['Name','Player_Score'])
for index ,row in goal_keepers.iterrows():
    ps = GK_Score(row['Name'],goal_keepers)
    new_entry = {'Name':row['Name'],'Player_Score':ps}
    GKList.loc[index] = new_entry
sorted_GKList = GKList.sort_values(by='Player_Score',ascending = False).drop_duplicates()
sorted_GKList.head(20)
```

```python
sorted_GKList['Name'].head(20).to_list()
```

# Top 20 Forwards

```python
ForwardsPlayer = player_data[player_data['Best Position'] == 'Forwards']
ForwardsList = pd.DataFrame(columns=['Name','Player_Score'])
for index ,row in ForwardsPlayer.iterrows():
    ps = Player_Score(row['Name'],ForwardsPlayer)
    new_entry = {'Name':row['Name'],'Player_Score':ps}
    ForwardsList.loc[index] = new_entry
sorted_ForwardsList = ForwardsList.sort_values(by='Player_Score',ascending = False).drop_duplicates()
sorted_ForwardsList.head(20)
```
